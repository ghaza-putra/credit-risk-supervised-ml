# Data Cleaning

## Mengubah Type Data

```python
#Mengubah Type Data ke Datetime
date_cols = ['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='%b-%y', errors='coerce')
```

## Menghapus Missing Value

```python
#Menghapus Data yang Memiliki 100% Missing Value
df = df.dropna(axis=1, how='all')
```

```python
df.info()
```

```python
df.isnull().mean().sort_values(ascending=False) * 100
```

```python
#Membuang Data yang tidak dipakai
df = df.drop(columns=[
    'desc', 'mths_since_last_record', 'next_pymnt_d',
    'mths_since_last_major_derog', 'url', 'emp_title',
    'title', 'zip_code', 'member_id', 'id', 'addr_state', 'mths_since_last_delinq'
])
```

```python
df.info()
```

```python
df['emp_length'].value_counts()
```

```python
# Mengonversi nilai emp_length dari string ke numerik (ordinal encoding)
def convert_emp_length(val):
    if pd.isnull(val) or val == 'n/a':
        return np.nan
    elif val == '< 1 year':
        return 0
    elif val == '10+ years':
        return 10
    else:
        return int(val.split()[0])

df['emp_length'] = df['emp_length'].apply(convert_emp_length)
```

```python
df['emp_length'].value_counts(dropna=False).sort_index()
```

```python
df.isnull().mean().sort_values(ascending=False) * 100
```

```python
#Mengisi Missing Value dengan Median
median_cols = [
    'total_rev_hi_lim',
    'tot_cur_bal',
    'tot_coll_amt',
    'revol_util',
    'inq_last_6mths',
    'total_acc',
    'pub_rec',
    'delinq_2yrs',
    'acc_now_delinq',
    'open_acc',
    'annual_inc',
    'collections_12_mths_ex_med',
    'emp_length'
]

for col in median_cols:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)
```

```python
#Mengisi Missing Value dengan Modus
datetime_cols = ['last_pymnt_d', 'last_credit_pull_d', 'earliest_cr_line']
for col in datetime_cols:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

Binary Encoding

```python
# Term: ubah menjadi 0 = 36 months, 1 = 60 months
df['term_bin'] = df['term'].str.extract('(\d+)').astype(int)
df['term_bin'] = df['term_bin'].apply(lambda x: 1 if x == 60 else 0)
```

```python
df['term_bin'].head(5)
```

```python
# Pymnt plan: n = 0, y = 1
df['pymnt_plan_bin'] = df['pymnt_plan'].map({'n': 0, 'y': 1})
```

```python
# Verification status: Not Verified = 0, lainnya = 1
df['verification_bin'] = df['verification_status'].apply(lambda x: 0 if x == 'Not Verified' else 1)
```

Ordinal Encoding

```python
# Grade: A-G -> 0–6
grade_order = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
df['grade_encoded'] = df['grade'].map({grade: i for i, grade in enumerate(grade_order)})
```

```python
# Loan Status: Fokus ke target → Fully Paid = 0, Charged Off + Default + Late = 1 (default risk)
loan_status_good = [
    'Fully Paid',
    'Does not meet the credit policy. Status:Fully Paid'
]  # Status pinjaman yang dianggap aman

loan_status_bad = [
    'Charged Off',
    'Default',
    'Late (31-120 days)',
    'Late (16-30 days)',
    'Does not meet the credit policy. Status:Charged Off'
] # Status pinjaman yang dianggap berisiko

df['loan_status_bin'] = df['loan_status'].apply(
    lambda x: 1 if x in loan_status_bad       # Jika status termasuk yang berisiko → beri label 1
    else (0 if x in loan_status_good          # Jika status termasuk yang aman → beri label 0
    else None)                                # Jika status tidak termasuk keduanya → beri None (NaN)
)

# Drop baris yang loan_status_bin = None (yang ambiguous)
df = df[df['loan_status_bin'].notnull()]
```

One-Hot Encoding

```python
df = df.drop([ 'sub_grade', 'policy_code'], axis=1)
```

#Data Preparation

```python
#Memilih Fitur untuk membantu prediksi
features = [
    'loan_amnt',                  # Jumlah pinjaman yang diajukan
    'installment',                # Cicilan bulanan pinjaman
    'int_rate',                   # Suku bunga pinjaman
    'emp_length',                 # Lama bekerja (dalam tahun)
    'annual_inc',                 # Pendapatan tahunan
    'dti',                        # Debt to income ratio (rasio utang terhadap penghasilan)

    'delinq_2yrs',                # Jumlah keterlambatan >30 hari dalam 2 tahun terakhir
    'inq_last_6mths',             # Jumlah pengajuan kredit dalam 6 bulan terakhir
    'open_acc',                   # Jumlah akun kredit yang aktif
    'pub_rec',                    # Catatan publik seperti kebangkrutan/gagal bayar
    'revol_bal',                  # Saldo utang bergulir (misalnya kartu kredit)
    'revol_util',                 # Rasio penggunaan kredit bergulir terhadap limit
    'total_acc',                  # Total akun kredit yang pernah dimiliki

    'out_prncp',                  # Sisa pokok utang yang belum dibayar
    'total_rec_late_fee',         # Denda yang pernah dibayar karena keterlambatan
    'recoveries',                 # Jumlah yang dipulihkan dari gagal bayar
    'collection_recovery_fee',    # Biaya pemulihan dari kolektor
    'last_pymnt_amnt',            # Jumlah pembayaran terakhir
    'collections_12_mths_ex_med', # Jumlah akun dalam koleksi (tidak termasuk medis)
    'acc_now_delinq',             # Akun yang sedang menunggak saat ini

    'tot_cur_bal',                # Total saldo seluruh akun kredit
    'total_rev_hi_lim',           # Batas maksimal kredit dari seluruh akun revolving
    'term_bin',                   # Lama tenor pinjaman (encoded)
    'grade_encoded'               # Grade pinjaman (A–G) yang di-encode
]
```

```python
X = df[features]
```

```python
#Memilih Label/Target yang akan diprediksi
y = df['loan_status_bin']
```

```python
#Melakukan Feature Scaling/Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train Test-Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

# Model & Evaluasi

```python
#Model Logistic Regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
#Prediksi
y_pred_lr = model.predict(X_test)
y_prob_lr = model.predict_proba(X_test)[:,1]
```

```python
#Evaluasi Model LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

print("Accuracy :", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr))
print("Recall   :", recall_score(y_test, y_pred_lr))
print("F1 Score :", f1_score(y_test, y_pred_lr))

# Confusion matrix
print(confusion_matrix(y_test, y_pred_lr))

# Laporan lengkap
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_lr)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Good', 'Bad'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix LR')
plt.show()

# Classification Report
print("Classification Report:\n")
print(classification_report(y_test, y_pred_lr, target_names=['Good', 'Bad']))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
auc_lr = roc_auc_score(y_test, y_prob_lr)
print(f"ROC AUC Score: {auc_lr:.4f}")

# Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob_lr)
plt.plot(fpr, tpr, label=f'LogReg AUC = {auc_lr:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Logistic Regression')
plt.legend()
plt.show()
```

```python
#Model Random Forest
from sklearn.ensemble import RandomForestClassifier

# Inisialisasi model dengan 100 pohon
rf_model = RandomForestClassifier(n_estimators=100, random_state=1)
rf_model.fit(X_train, y_train)

# Prediksi
y_pred_rf = rf_model.predict(X_test)
```

```python
#Evaluasi Model Random Forest
print("Accuracy:", rf_model.score(X_test, y_test))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, target_names=['Good', 'Bad']))
```

```python
# Tampilkan confusion matrix Random Forest secara visual
ConfusionMatrixDisplay.from_predictions(
    y_test, y_pred_rf,
    display_labels=['Good', 'Bad'],
    cmap='Blues'
)

plt.title('Confusion Matrix - Random Forest')
plt.grid(False)
plt.show()
```

```python
# Fitur yang berpengaruh terhadap model Random Forest
importances = rf_model.feature_importances_
feature_names = features

# Buat DataFrame untuk diurutkan
feat_importance = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Top 5 fitur
top5_rf = feat_importance.head(5)
print(top5_rf)

# Plot Grafik
sns.barplot(data=top5_rf, x='Importance', y='Feature')
plt.title("Top 5 Feature Importance - Random Forest")
plt.tight_layout()
plt.show()
```

```python
# # Fitur yang berpengaruh terhadap model Logistic Regression
coef = model.coef_[0]
feature_names = features

# Buat DataFrame
coef_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coef
}).sort_values(by='Coefficient', ascending=False)

# Top 5 dengan pengaruh positif
top5_pos = coef_df.head(5)

# Top 5 dengan pengaruh negatif
top5_neg = coef_df.tail(5)

print("Top 5 Coef (+):")
print(top5_pos)

print("\nTop 5 Coef (-):")
print(top5_neg)

# Plot Feature Importance untuk Logistic Regression (Koefisien)
plt.figure(figsize=(10, 6))
sns.barplot(x='Coefficient', y='Feature', data=top5_pos)
plt.title('Top 5 Fitur dengan Pengaruh Positif (Logistic Regression)')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Coefficient', y='Feature', data=top5_neg)
plt.title('Top 5 Fitur dengan Pengaruh Negatif (Logistic Regression)')
plt.show()
```
